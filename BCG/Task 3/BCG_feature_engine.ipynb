### Difference between off-peak prices in December and preceding January

```python
price_df = pd.read_csv('price_data.csv')
price_df["price_date"] = pd.to_datetime(price_df["price_date"], format='%Y-%m-%d')
price_df.head(3)
```

```python
price_df['price_date'].head()
```

```python
#client_df = pd.read_csv('client_data.csv')
#client_df.head(5)
```

```python
# Group off-peak prices by companies and month
monthly_price_by_id = price_df.groupby(['id', 'price_date']).agg({'price_off_peak_var': 'mean', 'price_off_peak_fix': 'mean'}).reset_index()

# Get january and december prices
jan_prices = monthly_price_by_id.groupby('id').first().reset_index()
dec_prices = monthly_price_by_id.groupby('id').last().reset_index()

# Calculate the difference
diff = pd.merge(dec_prices.rename(columns={'price_off_peak_var': 'dec_1', 'price_off_peak_fix': 'dec_2'}), jan_prices.drop(columns='price_date'), on='id')
diff['offpeak_diff_dec_january_energy'] = diff['dec_1'] - diff['price_off_peak_var']
diff['offpeak_diff_dec_january_power'] = diff['dec_2'] - diff['price_off_peak_fix']
diff = diff[['id', 'offpeak_diff_dec_january_energy','offpeak_diff_dec_january_power']]
display(diff.head())

# Select the 'offpeak_diff_dec_january_energy' and 'offpeak_diff_dec_january_power' columns from the DataFrame
energy_values = diff['offpeak_diff_dec_january_energy']
power_values = diff['offpeak_diff_dec_january_power']

# Create figure with two subplots
fig, axs = plt.subplots(ncols=2, figsize=(12, 5))

# Create histogram of 'offpeak_diff_dec_january_energy' values in first subplot
axs[0].hist(energy_values, bins=20)
axs[0].set_xlabel('Values')
axs[0].set_ylabel('Frequency')
axs[0].set_title('Histogram of Energy Data')

# Create histogram of 'offpeak_diff_dec_january_power' values in second subplot
axs[1].hist(power_values, bins=20)
axs[1].set_xlabel('Values')
axs[1].set_ylabel('Frequency')
axs[1].set_title('Histogram of Power Data')

# Display the plot
plt.show()
```

# Distribution transformation
From the previous EDA we can see that some features are highly skewed, we need to transform the distribution to normal-like distribution

```python
def apply_log_transform(df, columns):
    for col in columns:
        df.loc[df[col] < 0, col] = np.nan
        df[col].dropna(inplace=True)
        df[f"log_{col}"] = df[col].apply(lambda x: np.log(1 + x))
        _, FD_bins = np.histogram(df[f"log_{col}"], bins="fd")
        bin_nr = min(len(FD_bins)-1, 50)
        sns.displot(df[f"log_{col}"], bins=bin_nr, stat="density", alpha=0.4, kde=True, kde_kws={"cut": 3})

columns = ['cons_12m', 'cons_gas_12m', 'cons_last_month', 'imp_cons', 'forecast_cons_12m', 'forecast_cons_year', 'forecast_discount_energy', 'forecast_meter_rent_12m']
apply_log_transform(df, columns)
```

# Removing Outliers

```python
#I will use IQR method to detect outliers
def remove_outliers(df,col):
    df.loc[df[col] < 0,col]=df[col].mean()
    q1=df[col].quantile(.25)
    q3=df[col].quantile(.75)
    iqr= q3-q1
    upper_bound=q3+(iqr*1.5)
    lower_bound=q1-(iqr*1.5)
    df=df[(lower_bound<df[col])|(df[col]<upper_bound)]
```

```python
display(price_df.head())
display(df.head())
```

```python
columns_to_remove_outliers = ['mean_year_price_p1_var', 'mean_year_price_p2_var', 'mean_year_price_p3_var', 'mean_year_price_p1_fix', 'mean_year_price_p2_fix', 'mean_year_price_p3_fix', 'mean_year_price_p1', 'mean_year_price_p2', 'mean_year_price_p3', 'log_cons_12m', 'log_cons_gas_12m', 'log_cons_last_month',
                              'log_forecast_cons_12m', 'log_forecast_meter_rent_12m', 'log_forecast_cons_year', 'log_imp_cons', 'forecast_discount_energy', 'forecast_price_energy_p1', 'forecast_price_energy_p2', 'forecast_price_pow_p1', 'margin_gross_pow_ele', 'margin_net_pow_ele', 'net_margin', 'pow_max']

for column in columns_to_remove_outliers:
    if column in price_df.columns:
        remove_outliers(price_df, column)
    if column in df.columns:
        remove_outliers(df, column)
```
